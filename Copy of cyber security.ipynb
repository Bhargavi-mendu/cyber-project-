```python
text_ascii =[ca[ch] for ch in text]
key_ascii=[ca[ch] for ch in key]
```

```python
print(text,"=",text_ascii)
print(key,"=",key_ascii)
```

```python
x_enc=x.copy()
n=0
m=0
z=0
l=len(text)
kl=0
for i in range(1):
  org_val=x_enc[n,m,z]
  new_val=text_ascii[1]^key_ascii[(1%3)]
  x_enc[n,m,z]=new_val
  print("original value is ",org_val," after performing XOR operation at pixel (",n,m,z,") i.e., ",text_ascii[i]," XOR ",key_ascii[i%3]," the new value is ",new_val)
  n=n+1
  m=m+1
  m=(m+1)%3
  z=(z+1)%3
  kl=(kl+1)%len(key)
```
